## Extraemos texto con p.get_text()

```python
import requests
page = requests.get("https://feliperecabron.github.io/Trabajo_Nro1/")
page

page.content


from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')


body = list(html.children)[3]
list(body.children)
p = list(body.children)[1]
p.get_text()
```

## find_all nos revelara todas las etiquetas relacionadas, dependiendo de la que escojamos

```python
import requests
page = requests.get("https://feliperecabron.github.io/Trabajo_Nro1/")
page

page.content


from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')


soup.find_all('h2')
```

## soup.find_all('h2') + [3].get_text() nos permite buscar dentro de la etiqueta seleccionada la instancia que queramos Ej: h2, y dentro de h2 buscaremos la instancia que queramos podria ser el 0, el 1 o el 3 en este caso: 'Razas de Marihuana de mayor interes por las industria medicinales'

```python
import requests
page = requests.get("https://feliperecabron.github.io/Trabajo_Nro1/")
page

page.content


from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')


soup.find_all('h2')[3].get_text()
```
